<h1>Cleaning Dataset</h1>

```python
#Removing row - index :0 because it is empty
df = df.loc[1:]
df
```

```python
#Columns name are here at row-index: 1, so setting name of columns
df.columns = df.loc[1]
df
```

```python
# Now, we have no need of row-index-1 and row-index-2, so we are removing them
df = df.loc[3:]
df
```

```python
#Cleaning Data at last rows
df = df.loc[:632]
df
```

```python
#Upadating Columns Name
l = list(df.columns)
l.pop(0)
l.pop(0)
l.insert(0,'Month')
l.insert(0,'Year')
print(l)
```

```python
#Setting Columns
df.columns = l
df
```
