## Data Formatting

### Removing unnecessary double quotes from values

```python
for i in df['director']:
    print(i)
    break
```

```python
new_col = list()
col_list = ['rating_target_audience','genre','director','writer','star','country','company','gross']
for i in col_list:
    for j in df[i]:
        if type(j)==float:
            new_col.append(j)
        else:
            new_col.append(j[1:len(j)-1])
    df[i]=new_col
    new_col=[]
```

```python
df
```

```python
new_col2 =list()
col_list2 = ['release_date','release_year_and_country']
for i in col_list2:
    if i=='release_date':
        for j in df[i]:
            if type(j)==float:
                new_col2.append(j)
            else:
                new_col2.append(j[1:])
        df[i]=new_col2
        new_col2 =[]
    else:
        for j in df[i]:
            if type(j)==float:
                new_col2.append(j)
            else:
                new_col2.append(j[:len(j)-1])
        df[i]=new_col2
        new_col2 =[]
        
```

## Dealing With Missing Values

```python
df.isna().sum()
```

```python
df[['release_year_and_country','director','writer','star','country','company']]=df.loc[:,['release_year_and_country','director','writer','star','country','company']].replace(np.nan,'not mentioned')
```

```python
df.isna().sum()
```

**Replacing missing numerical data with 0**

```python
df[['score','votes','budget','gross','runtime']] = df.loc[:,['score','votes','budget','gross','runtime']].replace(np.nan,0)
```

```python
df
```
