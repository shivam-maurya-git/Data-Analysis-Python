```python
# Converting rainfall values into float

df[['1st', '2nd', '3rd', '4th', '5th', '6th',
       '7th', '8th', '9th', '10th', '11th', '12th', '13th', '14th', '15th',
       '16th', '17th', '18th', '19th', '20th', '21st', '22nd', '23rd', '24th',
       '25th', '26th', '27th', '28th', '29th', '30th', '31st']] = df[['1st', '2nd', '3rd', '4th', '5th', '6th',
       '7th', '8th', '9th', '10th', '11th', '12th', '13th', '14th', '15th',
       '16th', '17th', '18th', '19th', '20th', '21st', '22nd', '23rd', '24th',
       '25th', '26th', '27th', '28th', '29th', '30th', '31st']].astype('float')
```

```python
# Converting month numbering into float

df['month'] = df['month'].astype('float')
```

**Assigning Month Name to Month Numering**

```python
month_lookup = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
```

```python
month = list()
for i in df['month']:
    month_name = month_lookup[i]
    month.append(month_name)
df['month'] = month
```

```python
df.head()
```

```python
# Removing quotes from district values
district = list()
for i in df['district']:
    i = i[1:len(i)-1]
    district.append(i)
df['district'] = district
```
