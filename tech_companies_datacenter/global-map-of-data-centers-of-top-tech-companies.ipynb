## Importing libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

## Global Map of Meta's Datacenters

```python
meta = pd.read_csv('/kaggle/input/list-of-data-centers-of-top-tech-companies/meta_data_centers.csv')
```

```python
meta
```

```python
print('Hover on dots to see more infromation about datacenter and scroll towards left-right to see whole map')
print('DATA CENTERS OF META')
fig = px.scatter_mapbox(meta, lat="Latitude", lon="Longitude", hover_name="Location", hover_data=["Year Established", "Country",'Continent'],
                        color_discrete_sequence=["fuchsia"], zoom=1, height=700,title = 'Data centers of Meta')
fig.update_layout(mapbox_style="open-street-map")
fig.update_traces(marker=dict(size=10))
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

## Global Map of Amazon's Datacenters

```python
amazon = pd.read_csv('/kaggle/input/list-of-data-centers-of-top-tech-companies/amazon_data_centers.csv')
amazon.drop(1,inplace = True)
amazon.reset_index(inplace = True, drop = True)
amazon['Latitude'] = amazon['Latitude'].str.strip()
amazon['Longitude'] = amazon['Longitude'].str.strip()
amazon['Latitude'] = pd.to_numeric(amazon['Latitude'], errors='coerce')
amazon['Longitude'] = pd.to_numeric(amazon['Longitude'], errors='coerce')
amazon
```

```python
print('Hover on dots to see more infromation about datacenter and scroll towards left-right to see whole map')
print('DATA CENTERS OF Amazon')
fig = px.scatter_mapbox(amazon, lat="Latitude", lon="Longitude", hover_name="Location", hover_data=["Year Established", "Country",'Continent'],
                        color_discrete_sequence=["fuchsia"], zoom=1, height=700)
fig.update_layout(mapbox_style="open-street-map")
fig.update_traces(marker=dict(size=10))
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
