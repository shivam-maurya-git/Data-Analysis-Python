## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading Dataset

```python
df = pd.read_csv('/kaggle/input/netflix-ott-revenue-and-subscribers-csv-file/netflix_revenue_updated.csv')
```

```python
df
```

## Data Formatting : Checking Data types

```python
df.dtypes
```

```python
# Converting 'Date' column into date format
df['Date'] = pd.to_datetime(df['Date'],format = '%d-%m-%Y')
```

## Analyzing Data

```python
plt.figure(figsize = (8,5))
plt.plot(df['Date'],df['Global Revenue'])
plt.xlabel('Year-Month of Date')
plt.ylabel('Revenue in Billions US dollars')
plt.title('Netflix : Global Revenue in Billion US dollars')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.plot(df['Date'],df['LATM Streaming Revenue'],label ='LATM Streaming Revenue' ,color ='k')
plt.plot(df['Date'],df['UCAN Streaming Revenue'],label ='UCAN Streaming Revenue' ,color ='r')
plt.plot(df['Date'],df['APAC Streaming Revenue'],label ='UCAN Streaming Revenue' ,color ='b')
plt.plot(df['Date'],df['EMEA Streaming Revenue'],label ='EMEA Streaming Revenue' ,color ='g')
plt.xlabel('Year-Month of Date')
plt.ylabel('Revenue in Billions US dollars')
plt.title('Netflix : Region Wise Streaming Revenue in Billion US dollars')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.plot(df['Date'],df['Netflix Streaming Memberships '])
plt.xlabel('Year-Month of Date')
plt.ylabel('Number of Users in Hundreads Millions')
plt.title('Netflix : Global Netflix Streaming Memberships in Hundreads Millions')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.plot(df['Date'],df['LATM Members'],label ='LATM Members' ,color ='k')
plt.plot(df['Date'],df['UCAN Members'],label ='UCAN Members' ,color ='r')
plt.plot(df['Date'],df['APAC Members'],label ='UCAN Members' ,color ='b')
plt.plot(df['Date'],df['EMEA  Members'],label ='EMEA Members' ,color ='g')
plt.xlabel('Year-Month of Date')
plt.ylabel('Number of Streaming Membership in Millions')
plt.title('Netflix : Region Wise Streaming Membership in Ten Millions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.plot(df['Date'],df['LATM  ARPU'],label ='LATM ARPU' ,color ='k')
plt.plot(df['Date'],df['UCAN ARPU'],label ='UCAN ARPU' ,color ='r')
plt.plot(df['Date'],df['APAC  ARPU'],label ='APAC ARPU' ,color ='b')
plt.plot(df['Date'],df['EMEA ARPU'],label ='EMEA ARPU' ,color ='g')
plt.xlabel('Year-Month of Date')
plt.ylabel('Average Revenue per Users ($)')
plt.title('Netflix : Region Wise ARPU in US Dollars')
plt.legend()
plt.grid(True)
plt.show()
```
